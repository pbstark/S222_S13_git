```python
%pylab inline
import networkx as nx
import matplotlib.pyplot as plt
DG=nx.DiGraph()
DG.add_weighted_edges_from([(1,2,4), (2,1,3), (1,3,1), (3,4,1), (1,4,1), (1,4,1)])
nx.draw(DG)
```

```python
import scipy.sparse
A=scipy.sparse.eye(2,2,1)
DG=nx.from_scipy_sparse_matrix(A, create_using=nx.DiGraph())
nx.draw(DG)
```
