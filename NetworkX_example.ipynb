```python
import pickle
import networkx as nx
import matplotlib.pyplot as plt
matrix_comedy = open('matrix.pickle','r')
object_comedy = pickle.load(matrix_comedy)

object_comedy

```

```python
B = list()
for i in range(500):
    for j in range(500):
        if object_comedy[i,j]>0:
            B.append([i,j,object_comedy[i,j]])
B

```

```python

DG=nx.DiGraph()
DG.add_weighted_edges_from(B)
nx.draw(DG,node_size=0.1,edge_thickness=0.01)
plt.show()
```
